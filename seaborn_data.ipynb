# 🎨 Introduction to Seaborn

**Seaborn** is a powerful and easy-to-use data visualization library built on top of **Matplotlib** in Python. It provides a high-level interface for creating attractive and informative statistical graphics.

While Matplotlib handles the basic plotting, **Seaborn simplifies** the process by providing **built-in themes, color palettes**, and complex visualizations with minimal code.

---

## 🌟 Key Features of Seaborn

- Beautiful default styles and color themes
- Built-in support for **statistical plots**
- Works well with **Pandas DataFrames**
- Simplifies complex plots like:
  - **Distribution plots** (histograms, KDEs)
  - **Categorical plots** (boxplots, barplots, violin plots)
  - **Matrix plots** (heatmaps, correlation)
  - **Regression plots** (line of best fit with confidence intervals)

---

## ✅ Why Use Seaborn?

- Cleaner and more readable visualizations
- Quick comparisons and statistical summaries
- Great integration with data analysis tools like Pandas
- Saves time when creating complex graphs

## 🔥 Important Functions and Parameters for `sns.heatmap()`

- `data`: The 2D dataset to be visualized (NumPy array or pandas DataFrame).
- `annot=True`: Displays the numerical values inside the heatmap cells.
- `fmt="..."`: String formatting code to format annotation text (e.g., `".1f"` for 1 decimal place).
- `cmap="..."`: Color map to define the color scale (e.g., `"coolwarm"`, `"viridis"`, `"YlGnBu"`).
- `linewidths`: Width of the lines that divide each cell.
- `linecolor`: Color of the lines that divide each cell.
- `vmin`, `vmax`: Set the color scale's min and max values manually.
- `center`: The value at which to center the colormap when using a diverging palette.
- `xticklabels`, `yticklabels`: Show/hide axis tick labels (True, False, or list of labels).
- `square=True`: Makes the heatmap cells square-shaped.
- `mask`: Boolean array to mask certain cells (useful for upper/lower triangle).
- `cbar=True`: Show or hide the color bar.
- `cbar_kws={...}`: Dictionary of options to customize the color bar (e.g., `{"label": "Scale"}`).


````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create 2D data array, for example 4x5
data = np.random.rand(4, 5)

sns.heatmap(data, annot=True)
plt.show()

````

## 🎻 Important Functions and Parameters for `sns.violinplot()`

- `x`, `y`: Names of variables in `data` or vector data to be plotted on the respective axes.
- `data`: DataFrame containing the data to plot.
- `hue`: Grouping variable that will produce elements with different colors.
- `split=True`: When used with `hue`, draws split violins to compare distributions side-by-side.
- `inner`: Controls the display of the inner part of the violin (`"box"`, `"quartile"`, `"point"`, `"stick"`, or `None`).
- `scale`: The method used to scale the width of each violin (`"area"`, `"count"`, or `"width"`).
- `bw`: Bandwidth, which influences the smoothness of the violin (float between 0 and 1).
- `palette`: Color palette to be used for the plot.
- `order`: Order to plot the categorical levels on the axis.
- `linewidth`: Width of the outline of the violin.
- `cut`: Distance, in units of bandwidth, to extend the density past the extreme datapoints.
- `orient`: Orientation of the plot (`"v"` for vertical or `"h"` for horizontal).

📌 Use `plt.show()` at the end to display the plot in a script or Jupyter cell.

## 🔹 Sns `pairplot()` - Important Parameters

```python
sns.pairplot(
    data,               # DataFrame input
    hue=None,           # Group by color based on a categorical column
    palette="husl",     # Set color palette
    kind="scatter",     # Type of plot: 'scatter' or 'reg'
    diag_kind="kde",    # Diagonal plot: 'hist' or 'kde'
    markers=["o", "s"], # Marker styles for each level of hue
    height=2.5,         # Size of each subplot
    corner=False,       # If True, plot only lower triangle
    plot_kws={},        # Dict of keyword args for plots
    diag_kws={}         # Dict of keyword args for diagonal plots
)

## 🔹 Seaborn `stripplot()` – Important Parameters

```python

    data,            # DataFrame input
    x=None,          # Column name for x-axis
    y=None,          # Column name for y-axis
    hue=None,        # Grouping variable for color

## 🎨 Seaborn Plot – Common Parameters Reference

```python
sns.plot_function(
    data=df,              # 📊 The pandas DataFrame containing your data
    x="column_x",         # ➡️ Column name for x-axis
    y="column_y",         # ⬆️ Column name for y-axis
    hue="category",       # 🎨 Grouping variable (adds color distinction)
    style="style_col",    # 🧬 Marker/line style based on categories
    size="size_col",      # 🔘 Adjust marker size using a variable
    palette="Set2",       # 🌈 Choose color palette (e.g., "Set1", "coolwarm")
    markers=["o", "s"],   # 🔺 Marker styles (useful in scatter/line plots)
    kind="scatter",       # 📌 Type of plot (e.g., "scatter", "bar", "box", etc.)
    col="col_category",   # 🧱 Creates subplot columns (Faceting)
    row="row_category",   # 🧱 Creates subplot rows (Faceting)
    height=4,             # 📏 Height of each subplot (inches)
    aspect=1.2,           # ⚖️ Aspect ratio (width / height)
    dodge=True            # ↔️ Separate overlapping elements in categories
)

plt.show()
```

> 💡 **Tip:** Replace `sns.plot_function` with your specific plot like `sns.scatterplot`, `sns.barplot`, `sns.catplot`, etc.
