## Stack plot/ Area plot :

- x	The x-axis values (must be in ascending order)	x = [1, 2, 3, 4, 5]
- y1, y2, ...	Multiple y-series to stack vertically	y1 = [1, 2, 3, 4, 5]
- plt.stackplot()	The main function to create a stack plot	plt.stackplot(x, y1, y2, y3)
- labels	Labels for each y-series (used in legend)	labels=['Apple', 'Banana', 'Orange']
- colors	Colors for each stacked area	colors=['red', 'blue', 'green']
- plt.legend()	Displays the legend using the labels	plt.legend(loc='upper left')
- plt.title()	Sets the title of the graph	plt.title("Fruit Sales Over Time")
- plt.xlabel() / plt.ylabel()	Labels for x and y axes	plt.xlabel("Days"), plt.ylabel("Quantity")
- plt.show()	Displays the final plot	
-   **X values must be in accending order**


````python
import numpy as np
x =  [1, 2, 3, 4, 5]
y1 = [4, 5, 4, 7, 3]
y2 = [1, 2, 3, 4 ,5]
y3 = [2, 1, 2, 1, 2]

label = ["growth","sales","customer"]

plt.stackplot(x,y1,y2,y3,labels=label)
plt.legend(loc = 2,fontsize = 7)
plt.show()
````

## Step plot :

##  Step Plot in Matplotlib

A **Step Plot** is used to visualize discrete changes that happen at specific intervals. It is especially useful for showing how values change in **steps** rather than smooth transitions.

---

###  Main Elements Used in a Step Plot

| Element | Description | Example |
|--------|-------------|---------|
| `x` | The x-axis values | `x = [1, 2, 3, 4, 5]` |
| `y` | The y-axis values to plot as steps | `y = [3, 5, 2, 6, 4]` |
| `plt.step()` | The main function to draw a step plot | `plt.step(x, y)` |
| `where` | Defines how the steps should be positioned: `'pre'`, `'mid'`, `'post'` | `plt.step(x, y, where='mid')` |
| `color` | Line color of the step | `color='green'` |
| `label` | Label for the plot (used in legend) | `label='Sales Data'` |
| `plt.legend()` | Shows the label in the plot | `plt.legend()` |
| `plt.title()` | Title of the plot | `plt.title("Step Plot Example")` |
| `plt.xlabel()` / `plt.ylabel()` | Labels for axes | `plt.xlabel("Days")`, `plt.ylabel("Values")` |
| `plt.grid()` | Adds a grid to the plot | `plt.grid(True)` |
| `plt.show()` | Displays the plot |  |

---


````python
# x = [1,2,3,4,5]
# y = [1,2,1,3,7]
# y1 = [1,5,2,4,8]

# # l = ["python","C++"]
# plt.step(x,y,label="l")
# plt.xlabel("Days")
# plt.ylabel("Values")
# plt.title("Language")
# plt.legend()
# plt.grid()
# plt.show()
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 1, 3, 7]
y1 = [1, 5, 2, 4, 8]

plt.step(x, y, label="Python", where='mid')   # Step line for y
plt.step(x, y1, label="C++", where='mid')     # Step line for y1

plt.xlabel("Days")
plt.ylabel("Values")
plt.title("Language Progress")
plt.legend()
plt.grid(True)
plt.show()
````

## Fill between plot:

# `fill_between` Plot Explanation

`fill_between` is a function commonly used in plotting libraries (like Matplotlib in Python) to **fill the area between two curves or between a curve and a baseline** with color. This helps highlight specific regions on the graph.

---

## Basic Syntax

```python
plt.fill_between(x, y1, y2=0, color='color', alpha=transparency)


````python
# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = 0.5 * np.sin(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='0.5 * sin(x)')
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
plt.legend()
plt.show()
````
